# Validation of the Arc Characteristics Model for TIG Welding: 100% Helium

This Jupyter Notebook presents the validation of the results obtained with the `acm4tw` module, only for arcs modelled with a 100% Helium. For this purpose, diagrams comparing the calculated curves with those published by Delgado-lvarez et al. (2021) [[1](#ref-1)] and Delgado-lvarez et al. (2019) [[2](#ref-2)] were generated. Each diagram was created using the same axis scales as those reported in the published figures [[1](#ref-1), [2](#ref-2)] and subsequently superimposed using PowerPoint to allow a direct visual comparison.


## Boundary conditions of the modelled arcs

In the validation process, five different arc configurations were modelled, each defined by its own set of parameters (boundary conditions). These sets are listed below:
- Set 1: `gas_name = "Helium", current = 100 A, arclength = 4.5 mm`
- Set 2: `gas_name = "Helium", current = 150 A, arclength = 6.0 mm`
- Set 3: `gas_name = "Helium", current = 200 A, arclength = 8.0 mm`
- Set 4: `gas_name = "Helium", current = 250 A, arclength = 9.0 mm`
- Set 5: `gas_name = "Helium", current = 300 A, arclength = 10.0 mm`

Since the results do not show appreciable variations among arcs with different current intensities or arc lengths, only one representative case is presented in the following sections. This case corresponds to the *Set 3:* `"Helium", 200 A, 8.0 mm`.

It is important to note that, for every generated plot, the validity range of the arc shape must be enabled as follows:
- `valid_arc_shape = True`

The validity range of the arc column characteristics (`valid_arc_column`) does not apply in this case.

### Discussion: the temperature profile

The temperature profile for the case with 100% Helium is still computed with the general formulation for monoatomic gases (Equation 4 from Delgado-lvarez et al. 2021) [[1](#ref-1)]:

$$
\frac{T}{T_{\max}^\circ}\,\left(R,R_a(Z)\right)=a
+ b\left(\frac{R}{R_a}\mathrm{Pr}^{-1}\right)
+ c\left(\frac{R}{R_a}\mathrm{Pr}^{-1}\right)^2
+ d\left(\frac{R}{R_a}\mathrm{Pr}^{-1}\right)^3
+ e\left(\frac{R}{R_a}\mathrm{Pr}^{-1}\right)^4
+ f\left(\frac{R}{R_a}\mathrm{Pr}^{-1}\right)^5
+ g\left(\frac{R}{R_a}\mathrm{Pr}^{-1}\right)^6
$$

Where *a*, *b*, *c*, *d*, *e*, *f* and *g* are the coefficients extracted from the **Table 5** [[1](#ref-1)]. Each local maxima can be computed as a function of the axial position $Z$ by:

$$
\boxed{{Physical\ quantity}_{max}^\circ(Z) = a + bI + cL + dZ + (e \cdot {Physical\ quantity}_{max})^f}
$$

where $I$ is the current intensity, $L$ is the arc length and ${Physical\ quantity}_{max}$ is the global maxima for that property along all the arc column, computed as:

$$
\boxed{{Physical\ quantity}_{max} = a+bI+cL}
$$

The shape of the temperature profile presents some discrepances with the reference diagram. The reason for this could not be determined, but the hypothesis is that the constants to calculate the temperature profile must be different from those used to calculate the temperature with 100% Argon.

The Equation 4 [[1](#ref-1)] has its range of validity for all $(R/R_a)\mathrm{Pr}^{-1} \leq 1.0$. It was decided to accept the discrepancy with the profile for values of $(R/R_a)\mathrm{Pr}^{-1} > 1.0$

The profile is more similar to the one presented in the Figure 3b [[2](#ref-2)] for an arc modelled with a 100% Argon shielding gas, as shown in the Image 5:

### Discussion: the shear stress

The shear stress is only available in the `acm4tw` module for a shielding gas composition of 100% Argon, since no corresponding equation is provided in Delgado-lvarez et al. (2021) [[1](#ref-1)]. For this reason, in this Jupyter Notebook no shear stress is validated.

## Conclusion

All arc characteristics presented in this Jupyter Notebook show good agreement with the published diagrams.

The arc characteristic profiles are modelled using a specific equation for each physical quantity. All formulations rely on the same set of constants, which are extracted from the **Table 5** [[1](#ref-1)] and the **Table 2** [[2](#ref-2)]. As a consequence, the normalized profiles expressed as functions of the normalized arc radius always exhibit the same shape, regardless of the parameters used to model the arc column.

For the case of a 100% Helium shielding gas, the temperature profile follows a different profile than the ones showed in Delgado-lvarez et al. (2021). The hypothesis of this behaviour has already been discussed in the *Arc Characteristics Model for TIG Welding* Jupyter Notebook.
