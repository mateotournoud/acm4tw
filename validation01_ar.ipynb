# Validation of the Arc Characteristics Model for TIG Welding: 100% Argon

This Jupyter Notebook presents the validation of the results obtained with the `acm4tw` module, only for arcs modelled with a 100% Argon. For this purpose, diagrams comparing the calculated curves with those published by Delgado-lvarez et al. (2021) [[1](#ref-1)] and Delgado-lvarez et al. (2019) [[2](#ref-2)] were generated. Each diagram was created using the same axis scales as those reported in the published figures [[1](#ref-1), [2](#ref-2)] and subsequently superimposed using PowerPoint to allow a direct visual comparison.


## Boundary conditions of the modelled arcs

In the validation process, five different arc configurations were modelled, each defined by its own set of parameters (boundary conditions). These sets are listed below:
- Set 1: `gas_name = "Argon", current = 100 A, arclength = 4.5 mm`
- Set 2: `gas_name = "Argon", current = 150 A, arclength = 6.0 mm`
- Set 3: `gas_name = "Argon", current = 200 A, arclength = 8.0 mm`
- Set 4: `gas_name = "Argon", current = 250 A, arclength = 9.0 mm`
- Set 5: `gas_name = "Argon", current = 300 A, arclength = 10.0 mm`

Since the results do not show appreciable variations among arcs with different current intensities or arc lengths, only one representative case is presented in the following sections. This case corresponds to the *Set 3:* `"Argon", 200 A, 8.0 mm`.

It is important to note that, for every generated plot, the validity range of the shear stress must be enabled as follows:
- `valid_arc_shape = True`

The validity range of the arc column characteristics (`valid_arc_column`) does not apply in this case.

### Discussion: the temperature profile

The temperature profile is no longer computed using the general formulation for monoatomic gases. Instead, Equation 4 from Delgado-lvarez et al. (2019) [[2](#ref-2)] is employed for a shielding gas composition of 100% Argon, with a validity range of $R/R_a  1.5$:

$$
\frac{T}{T_{\max}^\circ}\,\left(R,R_a(Z)\right)=a
+ b\left(\frac{R}{R_a}\right)^2 + c\left(\frac{R}{R_a}\right)^4
+ d\left(\frac{R}{R_a}\right)^6 + e\left(\frac{R}{R_a}\right)^8
+ f\left(\frac{R}{R_a}\right)^{10} + g\left(\frac{R}{R_a}\right)^{12}
+ h\left(\frac{R}{R_a}\right)^{14} + i\left(\frac{R}{R_a}\right)^{16}
+ j\left(\frac{R}{R_a}\right)^{18}
$$

Where *a*, *b*, *c*, *d*, *e*, *f*, *g*, *h*, *i* and *j* are the coefficients extracted from the **Table 2** [[2](#ref-2)]. And the local maxima of the temperature at each axial position is given by the equation 7 [[2](#ref-2)] for $Z/L < 0.3$ or the equation 8 [[2](#ref-2)] for $Z/L  0.3$, respectively:

$$
T_{\max}^{\circ} = \left[a + b \left( \frac{z}{(R_c L)^{0.5}} \right)^{0.5}
- c \left( \frac{Z}{(R_c L)^{0.5}} \right)
+ d \left( \frac{Z}{(R_c L)^{0.5}} \right)^{1.5}
- e \left( \frac{Z}{(R_c L)^{0.5}} \right)^{2}
+ f \left( \frac{Z}{(R_c L)^{0.5}} \right)^{2.5}
- g \left( \frac{Z}{(R_c L)^{0.5}} \right)^{3}\right]\,*\, T_{\max}
$$

$$
T_{\max}^{\circ} = \left[a + b \ln\!\left( \frac{Z}{L} \right)
+ \frac{c}{\ln\!\left( \frac{Z}{L} \right)}
+ d \left( \ln\!\left( \frac{Z}{L} \right) \right)^2
+ \frac{e}{\left( \ln\!\left( \frac{Z}{L} \right) \right)^2}
+ f \left( \ln\!\left( \frac{Z}{L} \right) \right)^3
+ \frac{g}{\left( \ln\!\left( \frac{Z}{L} \right) \right)^3}
+ h \left( \ln\!\left( \frac{Z}{L} \right) \right)^4
\right]\,*\,\left(\frac{R_c^{0.5} \, I^{0.1}}{L^{0.2}}\right)
$$

The use of these equations was the only approach that produced a reasonable and valid temperature distribution, as shown in *Image 3*.

### Discussion: the shear stress

The shear stress is only available in the `acm4tw` module for a shielding gas composition of 100% Argon, since no corresponding equation is provided in Delgado-lvarez et al. (2021) [[1](#ref-1)]. For this reason, Equation 7 from Delgado-lvarez et al. (2019) [[2](#ref-2)] was adopted.

## Conclusion

All arc characteristics presented in this Jupyter Notebook show good agreement with the published diagrams.

The arc characteristic profiles are modelled using a specific equation for each physical quantity. All formulations rely on the same set of constants, which are extracted from the **Table 5** [[1](#ref-1)] and the **Table 2** [[2](#ref-2)]. As a consequence, the normalized profiles expressed as functions of the normalized arc radius always exhibit the same shape, regardless of the parameters used to model the arc column.

For the case of a 100% Argon shielding gas, the temperature profile follows a different formulation than those corresponding to 100% Helium or 50% Ar50% He gas mixtures. The reasons for this distinction have already been discussed in the *Arc Characteristics Model for TIG Welding* Jupyter Notebook.
